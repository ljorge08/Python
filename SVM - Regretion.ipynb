## SVM Regresi√≥n

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.sort(5*np.random.rand(200, 1), axis =0)
Y = np.sin(X).ravel()
Y[::5] += 3*(0.5 - np.random.rand(40))
```

```python
plt.scatter(X, Y, color="darkorange", label ="data")
```

```python
from sklearn.svm import SVR
```

```python
C = 1e3
svr_lin = SVR(kernel="linear", C=C)
svr_rbf = SVR(kernel="rbf", C=C, gamma=0.1)
svr_pol = SVR(kernel="poly", C=C, degree=1)
```

```python
y_lin = svr_lin.fit(X, Y).predict(X)
y_rbf = svr_rbf.fit(X, Y).predict(X)
y_pol = svr_pol.fit(X, Y).predict(X)
```

```python
lw = 2
plt.figure(figsize=(16, 9))
plt.scatter(X, Y, color="darkorange", label ="data")
plt.plot(X, y_lin, color="navy", label ="SVM Lineal", lw =lw)
plt.plot(X, y_rbf, color="c", label ="SVM Radial", lw =lw)
plt.plot(X, y_pol, color="cornflowerblue", label ="SVM Polin√≥mico", lw =lw)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Support Vector Regretion")
plt.legend()
plt.show()
```
