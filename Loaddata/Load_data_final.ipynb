```python
#En este ejemplo vemos el uso de la libreria urllib3, r.status dice el estado http de la página ideal para hacer alarmas y r.data la info de la URL 
#La respuesta del r.data es tipo byte, por eso se tiene que hacer un cast
def downloadFromURL(url, filename):
    import urllib3

    http = urllib3.PoolManager()
    r = http.request('GET', url)
    r.status
    r.data
    response = r.data
    response = str(response, 'utf-8')
    #print(type(response))

    allcolumns = {}

    response = response.strip().split("\n")
    #for column in response:

    encabezado = response[0]
    encabezado = encabezado.strip().split(",")
    lenenc = len(encabezado)

    data_frame = {}
    for col in encabezado:
        data_frame[col] = []

    counter=0

    for column in response:
        if(counter>0):
            columns = column.strip().split(",")
            for j in range(len(encabezado)):
                data_frame[encabezado[j]].append(columns[j]) 
        counter += 1        

    data_frame = pd.DataFrame(data_frame)
    data_frame.head()
    
    mainpath_xls = "C:/Users/l_jor/OneDrive/Documents/GitHub/python-ml-course/datasets"

#    filename = "/atlethes/atlethes_downloaded"
 
    data_frame.to_csv(mainpath_xls + filename + ".csv")
    data_frame.to_excel(mainpath_xls + filename + ".xls")
    data_frame.to_json(mainpath_xls + filename + ".json")
        
    return data_frame

#Acá llamo la función

medals_df=downloadFromURL(medals_urls, "/athletes/athletes_downloaded")
medals_df.head()
    
```
