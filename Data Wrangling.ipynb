```python
# Generar un numero aleatorio entre 0 a 1
np.random.random()
```

```python
# FunciÃ³n que genera una lista de numeros aleatorios entre el rango a y b
def randint_list(n,a,b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a, b))
    return x

randint_list(10,1,10)
```

```python
# La funcion randint_list ya hace lo descrito en la funciÃ³n anterior
randint_list(10, 1, 10)
```

```python
#CCON EL PAQUETE RANDOM FUNCION randrange SE PUEDE SACAR UN random en un rango multiplos de otro numero tercer argumento
import random
# Multiplos de 7 
random.randrange(0, 100, 7)
```

```python
# Si empiezo con 1 saca multiplos de 7 + 1
random.randrange(1, 100, 7)
```
